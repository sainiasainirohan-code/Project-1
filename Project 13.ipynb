```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Days"] = (df["Order Date"] - df["Order Date"].min()).dt.days
```

```python
features = ["Days", "Quantity", "Discount", "Category", "Sub-Category"]
X_encoded = pd.get_dummies(df[features], drop_first=True)
```

```python
X_train_rf, X_test_rf, Y_train_rf, Y_test_rf = train_test_split(X_encoded, Y, test_size=0.2, random_state=42)
```

```python
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train_rf, Y_train_rf)
```

```python
rf_pred = rf.predict(X_test_rf)
```

```python
r2_score(Y_test_rf, rf_pred)
```
